## TEST WAVE

```python
import yfinance as yf
from scipy.fft import fft, ifft, fftfreq, dct, idct
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
stock = yf.Ticker('eth-usd')
```

```python
db = stock.history(period='5y', interval='1d', group_by='ticker')
```

```python
val = db['Close']
```

```python
plt.plot(val)
```

### Define trenino

```python
val_today = val[-50:].to_numpy()
val_yest = val[-51:-1].to_numpy()

val_m2 = val[-52:-2].to_numpy()
print(str(val_today.shape)+' vs '+ str(val_yest.shape))
```

### DCT Forward

```python
# nSamples
N = len(val_today)
fy_today = dct(val_today)
fy_yest = dct(val_yest)
fy_m2 = dct(val_m2)
```

```python
#Take just 50 freqs
filt = np.zeros(len(fy_today))
filt[:60] = 1

iYest = idct(fy_yest)
iToday = idct(fy_today*filt)
err_today = []; err_yest = []
for i in range(len(iYest)):
    err_today.append(abs(val_today[i] - iToday[i]))
    err_yest.append(abs(val_yest[i] - iYest[i]))
```

```python
plt.plot(val[-50:].to_list())
plt.plot(iToday, 'k')
plt.plot(iYest[1:], 'r')
```

```python
plt.plot((fy_m2*filt)[1:],'g')
plt.plot((fy_yest*filt)[1:],'r')
plt.plot((fy_today*filt)[1:],'k')
```
